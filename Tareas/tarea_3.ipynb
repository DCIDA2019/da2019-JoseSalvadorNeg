# Visualización con Matplotlib #

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
#Librerías que se utilizarán en la tarea#
```

```python
plt.style.use('classic')
#Estilo clásico de plot#
```

Dependiendo del entorno de trabajo, existen distintas formas de utilizar matplotlib.

## matplotlib desde un "código" ##

El uso de matplotlib para un "código" (script) es tan sencillo como hacer uso de la función plt.show(). Un ejemplo sencillo es tener un archivo del tipo .py que contenga un código del estilo

```python
import numpy as np
x = np.linspace(0, 5, 50)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()

#Correr este código dará como resultado una gráfica como la mostrada a continuación#
```

## matplotlib desde una IPython shell ##

Es posible usar matplotlib en IPython si se especifica el modo matplotlib. Basta con hacer uso del comando "mágico" %matplotlib una vez que se inicie IPython.

## matplotlib desde una IPython notebook ##

Las notebook de IPython se caracterizan por ser flexibles a la hora de cargar con elementos de todo tipo (código, comentarios, html, imágenes, etc.), lo que la vuelve una interfaz interactiva y simple (este archivo es una notebook). De manera similar a las shell, el uso de matplotlib es tan sencillo como llamar a la función %matplotlib para así poder comenzar a graficar. A su vez existen dos tipos de uso de %matplotlib: notebook e inline (la primera genera gráficas interactivas para el notebook, mientras que la segunda genera gráficas estáticas y sencillas).

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 50)

fig = plt.figure()
#'-' nos dará una línea continua, mientras que '--' una punteada#
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

## Interfaces ##

### Interfaz de estilo MATLAB ###

Al ser matplotlib una alternativa de uso con respecto a MATLAB, es de esperar que matplotlib mantuviera similitudes con MATLAB. Lo siguiente es un ejemplo de ello

```python
plt.figure #creamos una gráfica#

#creación del primer panel (fila, columna, no. de panel)#
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

#creación del segundo panel (fila, columna, no. de panel)#
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

### Interfaz orientada a objetos ###

Esta interfaz permite mayor control sobre las gráficas al otorgar herramientas para elegir el eje de la gráfica, así como la figura

```python
#Creamos un conjunto de gráficas#
#eje será el arreglo de dos objetos clasificados como Ejes#
fig, eje = plt.subplots(2)

#Hacemos uso de la función objeto, con su respectivo eje#
eje[0].plot(x, np.sin(x))
eje[1].plot(x, np.cos(x));
```

# Gráficas de líneas simples #

Las gráficas más sencillas son aquellas que representan líneas simples (algo de la forma y = f(x)). Este sería el uso más sencillo de la herramienta matplotlib.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
import pandas as pd
tabulador = pd.read_excel('tabulador_pe2019.xls', header=9) #datos inician en el renglón 10#
tabulador
```

```python
print(tabulador.columns)
tabulador.head(2)
```

```python
tabulador['SUELDO MENSUAL BRUTO'].unique()
```

```python
#El siguiente es un ejemplo de ploteo sencillo#
tabulador.plot(kind = 'line', x = 'NIVEL', y = 'SUELDO BASE')
plt.show()
```

```python
#También es posible colocar dos funciones en la misma figura#
ax = plt.gca()

tabulador.plot(kind = 'line', color = 'green', x = 'NIVEL', y = 'SUELDO BASE', ax = ax)#Elección de color verde mediante su nombre en inglés#
tabulador.plot(kind = 'line', color = 'y', x = 'NIVEL', y = 'APOYO FAMILIAR', ax = ax)#Elección de color amarillo mediante su abreviación con la letra y#
plt.show()
```

```python
ax = plt.gca()

tabulador.plot(kind = 'line', color = '0.5', x = 'NIVEL', y = 'SUELDO BASE', ax = ax)#Elección de color mediante escala de grises (del 0 al 1)#
tabulador.plot(kind = 'line', color = '#e4007c', x = 'NIVEL', y = 'APOYO FAMILIAR', ax = ax)#Elección de color rosa mexicano mediante código hex#
plt.show()
```

```python
ax = plt.gca()

tabulador.plot(kind = 'line', color = (0.5, 1.0, 0.5), x = 'NIVEL', y = 'SUELDO BASE', ax = ax)#Elección de color mediante escala RGB (del 0 al 1)#
tabulador.plot(kind = 'line', color = 'indigo', x = 'NIVEL', y = 'APOYO FAMILIAR', ax = ax)#Colores que se encuentren en código HTML#
plt.show()
```

No especificar el color, hace que se asigne un color predeterminado.

```python
ax = plt.gca()

tabulador.plot(kind = 'line', linestyle = '-', x = 'NIVEL', y = 'SUELDO BASE', ax = ax)#Línea normal#
tabulador.plot(kind = 'line', linestyle = '--', x = 'NIVEL', y = 'APOYO FAMILIAR', ax = ax)#Líneas separadas#
tabulador.plot(kind = 'line', linestyle = '-.', x = 'NIVEL', y = 'GRATIFICACIÓN QUINCENAL', ax = ax)#Línea-punto#
tabulador.plot(kind = 'line', linestyle = ':', x = 'NIVEL', y = 'SUELDO MENSUAL NETO', ax = ax)#Línea de puntos#
plt.show()
```

A su vez es posible simplificar todo mediante una notación de tipo de línea-color (por ej. si queremos una línea de puntos roja, ponemos ':r' en el ploteo).

Es también posible delimitar los ejes en lugar de dejar que la máquina lo haga.

```python
tabulador.plot(kind = 'line', linestyle = '--', x = 'NIVEL', y = 'APOYO FAMILIAR')

plt.xlim(1, 10)
plt.ylim(5000, 10000);
```

```python
tabulador.plot(kind = 'line', linestyle = '--', x = 'NIVEL', y = 'APOYO FAMILIAR')

plt.xlim(10, 1)
plt.ylim(10000, 5000); #Esto invierte los ejes#
```

Una forma alternativa de establecer límites en los ejes consiste en el uso de la función plt.axis. A su vez dicha función es capaz de ajustar la tabla para colocar los límites en donde tengamos puntos.

Por último, existe una forma de colocar nombres a nuestras figuras.

```python
tabulador.plot(kind = 'line', linestyle = '--', x = 'NIVEL', y = 'APOYO FAMILIAR', label = 'Esta es una línea')
plt.title("Esta es una figura")
plt.xlabel("Este es el eje x")
plt.ylabel("Este es el eje y")
```

## Ploteo de puntos dispersos simples ##

Parecido al ploteo con líneas, solo que aquí, usamos puntos.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
tabulador.plot(kind = 'scatter', linestyle = '--', x = 'NIVEL', y = 'APOYO FAMILIAR')
```

```python
#Existen diferentes formas de representar puntos en una figura#
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```
