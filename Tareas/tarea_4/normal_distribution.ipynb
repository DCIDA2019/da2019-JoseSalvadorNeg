Find the cumulative distribution function for the gaussian distribution and reproduce the plots. Choose a random number between 0 and 1 and use the corresponding value of H0. Generate as many as you want and make the histogram of H0 to verify you did it right. Use a mean of 70 and a sigma of 2.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats
import scipy.special
import random
plt.style.use('classic')
%matplotlib inline
```

```python
mean = 70
sigma = 2

x = np.linspace(60, 80, 100)
y = scipy.stats.norm.pdf(x, mean, sigma)

plt.plot(x, y, '-', color = 'indigo', linewidth = 3)
plt.ylim(0, 0.22)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title('Gaussian distribution')
```

```python
x = np.linspace(60, 80, 100)
y = (1/2)*(1 + scipy.special.erf((x - mean)/(sigma*np.sqrt(2))))

plt.plot(x, y, '.', color = 'indigo')
plt.ylim(-0.1, 1.1)
plt.xlabel('x')
plt.ylabel('cdf')
plt.title('Cumulative distribution function')
```

```python
H = []
for i  in range(10000):
    ran = random.random()
    H.append(mean + sigma*np.sqrt(2)*(scipy.special.erfinv(2*ran - 1)));


plt.hist(H, bins = 100, color = 'indigo')
```

```python
x = np.linspace(60, 80, 100)
y = scipy.stats.norm.pdf(x, mean, sigma)

plt.plot(x, y, '-', color = 'orangered', linewidth = 3)
plt.ylim(0, 0.22)
plt.hist(H, bins = 100, color = 'indigo', density = True)
```
