# Statistics and Probability #

For the distributions mentioned, find the CDF, mean and standard deviation. Plot both the PDF and CDF, for some different values of mean and sigma. Use a random number generator to compute sample the PDF and compare it with the CDF using the corresponding Python function.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats
import scipy.special
import random
import math
plt.style.use('classic')
%matplotlib inline
```

## Poison distribution ##

```python
lamb1 = 1
lamb2 = 5
lamb3 = 10

x1 = np.linspace(1, 100, 100)
y1 = scipy.stats.poisson.pmf(x1, lamb1, lamb1)

x2 = np.linspace(1, 100, 100)
y2 = scipy.stats.poisson.pmf(x2, lamb2, lamb2)

x3 = np.linspace(1, 100, 100)
y3 = scipy.stats.poisson.pmf(x3, lamb3, lamb3)

plt.plot(x1, y1, '-', color = 'indigo', linewidth = 3)
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 3)
plt.plot(x3, y3, '-', color = 'lime', linewidth = 3)
plt.xlabel('x')
plt.ylabel('pmf')
plt.title('Poisson distribution')
```

```python
x1 = np.linspace(1, 100, 100)
y1 = scipy.stats.poisson.cdf(x1, lamb1, lamb1)

x2 = np.linspace(1, 100, 100)
y2 = scipy.stats.poisson.cdf(x2, lamb2, lamb2)

x3 = np.linspace(1, 100, 100)
y3 = scipy.stats.poisson.cdf(x3, lamb3, lamb3)

plt.plot(x1, y1, '-', color = 'indigo', linewidth = 3)
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 3)
plt.plot(x3, y3, '-', color = 'lime', linewidth = 3)
plt.xlabel('x')
plt.ylabel('cdf')
plt.title('Poisson distribution')
```

```python
plt.plot(y2, x2, '-', color = 'orangered', linewidth = 3)
plt.xlim(-0.05, 1.05)
```

```python
y2_new = np.random.uniform(0.05, 0.95, 10000)
f = scipy.interpolate.interp1d(y2, x2)
x2_new = f(y2_new)
plt.plot(y2_new, x2_new, '-', color = 'lime', linewidth = 3)
plt.plot(y2, x2, '-', color = 'orangered', linewidth = 2)
plt.xlim(-0.05, 1.05)
```

```python
x2 = np.linspace(1, 100, 100)
y2 = scipy.stats.poisson.pmf(x2, lamb2, lamb2)
plt.hist(x2_new, bins = 10, density = True, color = 'indigo')
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 2)
```

## Beta distribution ##

```python
alfa1 = 0.5
alfa2 = 1.0
alfa3 = 5.0
beta1 = 0.5
beta2 = 3.0
beta3 = 1.0

x1 = np.linspace(0, 1, 100)
y1 = scipy.stats.beta.pdf(x1, alfa1, beta1)

x2 = np.linspace(0, 1, 100)
y2 = scipy.stats.beta.pdf(x2, alfa2, beta2)

x3 = np.linspace(0, 1, 100)
y3 = scipy.stats.beta.pdf(x3, alfa3, beta3)

plt.plot(x1, y1, '-', color = 'indigo', linewidth = 3)
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 3)
plt.plot(x3, y3, '-', color = 'lime', linewidth = 3)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title('Beta distribution')
```

```python
x1 = np.linspace(0, 1, 100)
y1 = scipy.stats.beta.cdf(x1, alfa1, beta1)

x2 = np.linspace(0, 1, 100)
y2 = scipy.stats.beta.cdf(x2, alfa2, beta2)

x3 = np.linspace(0, 1, 100)
y3 = scipy.stats.beta.cdf(x3, alfa3, beta3)

plt.plot(x1, y1, '-', color = 'indigo', linewidth = 3)
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 3)
plt.plot(x3, y3, '-', color = 'lime', linewidth = 3)
plt.xlabel('x')
plt.ylabel('cdf')
plt.title('Poisson distribution')
```

```python
plt.plot(y2, x2, '-', color = 'orangered', linewidth = 3)
plt.xlim(-0.05, 1.05)
```

```python
y2_new = np.random.uniform(0.05, 0.95, 10000)
f = scipy.interpolate.interp1d(y2, x2)
x2_new = f(y2_new)
plt.plot(y2_new, x2_new, '.', color = 'lime', linewidth = 3)
plt.plot(y2, x2, '-', color = 'orangered', linewidth = 2)
plt.xlim(-0.05, 1.05)
```

```python
x2 = np.linspace(0, 1, 100)
y2 = scipy.stats.beta.pdf(x2, alfa2, beta2)
plt.hist(x2_new, bins = 10, density = True, color = 'indigo')
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 2)
```

## Binomal distribution ##

```python
x1 = np.linspace(1, 50, 50)
y1 = scipy.stats.binom.pmf(x1, 50, 0.5)

x2 = np.linspace(1, 60, 60)
y2 = scipy.stats.binom.pmf(x2, 60, 0.8)

x3 = np.linspace(1, 70, 70)
y3 = scipy.stats.binom.pmf(x3, 70, 0.4)

plt.plot(x1, y1, '-', color = 'indigo', linewidth = 3)
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 3)
plt.plot(x3, y3, '-', color = 'lime', linewidth = 3)
plt.xlabel('x')
plt.ylabel('pmf')
plt.title('Binomial distribution')
```

```python
x1 = np.linspace(1, 50, 50)
y1 = scipy.stats.binom.cdf(x1, 50, 0.5)

x2 = np.linspace(1, 60, 60)
y2 = scipy.stats.binom.cdf(x2, 60, 0.8)

x3 = np.linspace(1, 70, 70)
y3 = scipy.stats.binom.cdf(x3, 70, 0.4)

plt.plot(x1, y1, '-', color = 'indigo', linewidth = 3)
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 3)
plt.plot(x3, y3, '-', color = 'lime', linewidth = 3)
plt.xlabel('x')
plt.ylabel('cdf')
plt.title('Poisson distribution')
```

```python
plt.plot(y2, x2, '-', color = 'orangered', linewidth = 3)
```

```python
y2_new = np.random.uniform(0.05, 0.95, 10000)
f = scipy.interpolate.interp1d(y2, x2)
x2_new = f(y2_new)
plt.plot(y2_new, x2_new, '.', color = 'lime', linewidth = 3)
plt.plot(y2, x2, '-', color = 'orangered', linewidth = 2)
plt.xlim(-0.05, 1.05)
```

```python
x2 = np.linspace(1, 60, 60)
y2 = scipy.stats.binom.pmf(x2, 60, 0.8)
plt.hist(x2_new, bins = 10, density = True, color = 'indigo')
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 2)
```

## Chi-squared distribution ##

```python
x1 = np.linspace(1, 10, 100)
y1 = scipy.stats.chi2.pdf(x1, 1)

x2 = np.linspace(1, 10, 100)
y2 = scipy.stats.chi2.pdf(x2, 3)

x3 = np.linspace(1, 10, 100)
y3 = scipy.stats.chi2.pdf(x3, 5)

plt.plot(x1, y1, '-', color = 'indigo', linewidth = 3)
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 3)
plt.plot(x3, y3, '-', color = 'lime', linewidth = 3)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title('Chi-squared distribution')
```

```python
x1 = np.linspace(1, 10, 100)
y1 = scipy.stats.chi2.cdf(x1, 1)

x2 = np.linspace(1, 10, 100)
y2 = scipy.stats.chi2.cdf(x2, 3)

x3 = np.linspace(1, 10, 100)
y3 = scipy.stats.chi2.cdf(x3, 5)

plt.plot(x1, y1, '-', color = 'indigo', linewidth = 3)
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 3)
plt.plot(x3, y3, '-', color = 'lime', linewidth = 3)
plt.xlabel('x')
plt.ylabel('cdf')
plt.title('Chi-squared distribution')
```

```python
plt.plot(y2, x2, '-', color = 'orangered', linewidth = 3)
```

```python
y2_new = np.random.uniform(0.2, 0.98, 10000)
f = scipy.interpolate.interp1d(y2, x2)
x2_new = f(y2_new)
plt.plot(y2_new, x2_new, '.', color = 'lime', linewidth = 3)
plt.plot(y2, x2, '-', color = 'orangered', linewidth = 2)
plt.xlim(-0.05, 1.05)
```

```python
x2 = np.linspace(1, 10, 100)
y2 = scipy.stats.chi2.pdf(x2, 3)
plt.hist(x2_new, bins = 10, density = True, color = 'indigo')
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 2)
```

## Additional distribution ##

### Pareto's distribution ###

```python
x1 = np.linspace(1, 3, 100)
y1 = scipy.stats.pareto.pdf(x1, 1)

x2 = np.linspace(1, 3, 100)
y2 = scipy.stats.pareto.pdf(x2, 5)

x3 = np.linspace(1, 3, 100)
y3 = scipy.stats.pareto.pdf(x3, 10)

plt.plot(x1, y1, '-', color = 'indigo', linewidth = 3)
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 3)
plt.plot(x3, y3, '-', color = 'lime', linewidth = 3)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title('Pareto distribution')
```

```python
x1 = np.linspace(1, 3, 100)
y1 = scipy.stats.pareto.cdf(x1, 1)

x2 = np.linspace(1, 3, 100)
y2 = scipy.stats.pareto.cdf(x2, 5)

x3 = np.linspace(1, 3, 100)
y3 = scipy.stats.pareto.cdf(x3, 10)

plt.plot(x1, y1, '-', color = 'indigo', linewidth = 3)
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 3)
plt.plot(x3, y3, '-', color = 'lime', linewidth = 3)
plt.xlabel('x')
plt.ylabel('cdf')
plt.title('Pareto distribution')
```

```python
plt.plot(y2, x2, '-', color = 'orangered', linewidth = 3)
```

```python
y2_new = np.random.uniform(0.0001, 0.9955, 10000)
f = scipy.interpolate.interp1d(y2, x2)
x2_new = f(y2_new)
plt.plot(y2_new, x2_new, '.', color = 'lime', linewidth = 3)
plt.plot(y2, x2, '-', color = 'orangered', linewidth = 2)
plt.xlim(-0.05, 1.05)
```

```python
x2 = np.linspace(1, 3, 100)
y2 = scipy.stats.pareto.pdf(x2, 5)
plt.hist(x2_new, bins = 100, density = True, color = 'indigo')
plt.plot(x2, y2, '-', color = 'orangered', linewidth = 2)
```

Pareto's distribution has many applications. The next examples are (sometimes) approximately Pareto-distributed:

- The size of meteorites
- Clusters of Bose-Einstein condensate near absolute zero
- Hard disk drive error rates
