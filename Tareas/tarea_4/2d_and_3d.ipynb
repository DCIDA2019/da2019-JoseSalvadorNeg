# Statistics and Probability #

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats
import scipy.special
import random
plt.style.use('classic')
%matplotlib inline
```

```python
r = np.random.uniform(0, 1, 2222)
theta = np.random.uniform(0, 2*np.pi, 2222)

plt.plot(r, theta, '.')
plt.xlabel('r')
plt.ylabel('theta')
plt.ylim(-0.1, 6.5)
```

```python
x = np.random.uniform(-1, 1, 2222)
y = np.random.uniform(-1, 1, 2222)

x_new = r*np.cos(theta)
y_new = r*np.sin(theta)

plt.plot(x, y, '.', color = 'indigo')
plt.plot(x_new, y_new, '.', color = 'orangered')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 1)
```

```python
x = np.random.uniform(-1, 1, 2222)
y = np.random.uniform(-1, 1, 2222)

x_aux = np.random.uniform(-1, 1, 2222)
y_aux = np.random.uniform(-1, 1, 2222)

r_aux = np.sqrt(x_aux**2 + y_aux**2)
r_new = []
theta_new = []

for i in range(2222):
    if r_aux[i] <= 1:
        r_new.append(r_aux[i])
        theta_new.append(np.arctan2(y_aux[i], x_aux[i]));

x_new = r_new*np.cos(theta_new)
y_new = r_new*np.sin(theta_new)

plt.plot(x, y, '.', color = 'indigo')
plt.plot(x_new, y_new, '.', color = 'orangered')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 1)
```

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')

r = np.random.uniform(0, 1, 555)
theta = np.random.uniform(0, np.pi, 555)
phi = np.random.uniform(0, 2*np.pi, 555)

x_new = r*np.sin(theta)*np.cos(phi)
y_new = r*np.sin(theta)*np.sin(phi)
z_new = r*np.cos(theta)

ax.scatter3D(x_new, y_new, z_new, 'og', c = z_new, cmap = 'Greens', linewidth = 0.5);
```

```python
x = np.random.uniform(-1, 1, 333)
y = np.random.uniform(-1, 1, 333)
z = np.random.uniform(-1, 1, 333)

x_aux = np.random.uniform(-1, 1, 333)
y_aux = np.random.uniform(-1, 1, 333)
z_aux = np.random.uniform(-1, 1, 333)

ax = plt.axes(projection='3d')

r_aux = np.sqrt(x_aux**2 + y_aux**2 + z_aux**2)
r_new = []
theta_new = []
phi_new = []

for i in range(333):
    if r_aux[i] <= 1:
        r_new.append(r_aux[i])
        theta_new.append(np.arctan2(y_aux[i], x_aux[i]))
        phi_new.append(np.arctan2(np.sqrt(x_aux[i]**2 + y_aux[i]**2), z_aux[i]));

x_new = r_new*np.sin(theta_new)*np.cos(phi_new)
y_new = r_new*np.sin(theta_new)*np.sin(phi_new)
z_new = r_new*np.cos(theta_new)

ax.scatter3D(x_new, y_new, z_new, 'og', c = z_new, cmap = 'Greens', linewidth = 0.5);
```
