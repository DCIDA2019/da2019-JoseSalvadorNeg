# Statistics and Probability #

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats
import scipy.special
import random
plt.style.use('classic')
%matplotlib inline
```

```python
# Distribución p(x) #

x = np.linspace(0, 1, 50)
y = np.ones(50)

plt.plot(x, y, '-', color = 'indigo', linewidth = 5)
```

```python
# Distribución p(y) #

x = np.linspace(1, np.exp(1), 50)
y = 1/x

plt.plot(x, y, '-', color = 'indigo', linewidth = 5)
```
