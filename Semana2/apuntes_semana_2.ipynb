```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f_1(x):
    if x%2 == 0:
        return x*x
    else:
        return np.log(x)
```

```python
xx = np.arange(1, 50, 3)
```

```python
def f_2(x):
    y = []
    for i in range (len(x)):
        if x[i]%2 == 0:
            y.append(x[i]*x[i])
        else:
            y.append(np.log(x[i]))
    return x,y
```

```python
f_3 = np.vectorize(f_1)
```

```python
f_3(xx)
```

```python
def g(x):
    if x > 0:
        return np.cos(x)
    else:
        return np.sin(x)

g1 = np.vectorize(g)
```

```python
x2 = np.linspace(-2*np.pi, 2*np.pi)
g1(x2)
```

```python
plt.plot(x2, g1(x2))
```

```python
y = np.zeros_like(x2)
w = x2 > 0
%timeit y[w] = g1(x2[w])
%timeit y[~w] = g1(x2[~w])

##segundo mejor método##
```

```python
y
```

```python
y2 = np.asarray([g(x) for x in x2])
%timeit y2

##mejor método##
```

```python
def g2(x):
    y = np.zeros_like(x)
    for i in range(len(x)):
        if x[i] > 0:
            np.cos(x[i])
        else:
            np.sin(x[i])
    return y
```

```python
%timeit g2(x2)

##peor método dado el tiempo que le toma##
```

```python
class funciones:
    
    def __init__(self, A, saludo = 'Howdy'):
        self.A = A
        self.saludo = 'Howdy'
        
    def g(self, x):
        return self.A*np.cos(x), self.A*np.sin(x)
    
    def f2(self, x):
        if x%2 == 0:
            return x*x
        else:
            return np.log(x)
    def f(self):
        return self.saludo
```

```python
x = funciones(6, 'Howdy')
```

```python
x.f()
```

```python
x.g(np.pi)
```
